```python
import pandas as pd
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Example importing the CSV here
cpu_train_a= 'https://raw.githubusercontent.com/oreilly-mlsec/book-resources/master/chapter3/datasets/cpu-utilization/cpu-train-a.csv'
cpu_train_b= 'https://raw.githubusercontent.com/oreilly-mlsec/book-resources/master/chapter3/datasets/cpu-utilization/cpu-train-b.csv'
cpu_test_a='https://raw.githubusercontent.com/oreilly-mlsec/book-resources/master/chapter3/datasets/cpu-utilization/cpu-test-a.csv'
cpu_test_b= 'https://raw.githubusercontent.com/oreilly-mlsec/book-resources/master/chapter3/datasets/cpu-utilization/cpu-test-b.csv'

cpu_train_a = pd.read_csv(cpu_train_a)
cpu_train_b = pd.read_csv(cpu_train_b)
cpu_test_a = pd.read_csv(cpu_test_a)
cpu_test_b = pd.read_csv(cpu_test_b)
cpu_train_a
```

```python
cpu_train_a.info()
```

```python
cpu_train_a['date'] = pd.to_datetime(cpu_train_a['datetime'])
cpu_train_a = cpu_train_a.set_index('date')
cpu_train_a.drop(['datetime'], axis=1, inplace=True)
cpu_train_a.head()
```

```python
cpu_train_a.info()
```

```python
cpu_train_b['date'] = pd.to_datetime(cpu_train_b['datetime'])
cpu_train_b = cpu_train_b.set_index('date')
cpu_train_b.drop(['datetime'], axis=1, inplace=True)
cpu_train_b.head()
```

```python
cpu_train_b.info()
```

```python
plt.plot(cpu_train_a['cpu'])
plt.show()
```

```python
plt.plot(cpu_train_b['cpu'])
plt.show()
```
